```python
# LINE GRAPH
a = [1,3,5,7,9,15]
b = (4,6,8,10,12,14)
c = np.array([1,9,12,18,21,24])
```

```python
plt.plot(a,b,color='#1E0BEB',label = 'Blue Line',linestyle = 'dashdot',linewidth = 3,alpha = 0.7)
plt.plot(b,c,color = '#FF0000',label = "Red Line",marker = "*",markerfacecolor="black",markeredgecolor="black",markersize=10)
plt.plot(a,c,color = '#86FF6D',label = "Green Line",linewidth = 5)
plt.title("COLORED LINES")
plt.xlabel("X - AXIS")
plt.ylabel("Y - AXIS")
plt.legend(title = "TYPES OF LINES")
plt.grid()
plt.show()

# linestyles - solid,dashed,dotted,dashdot
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
```

```python
#BAR GRAPH
a = ["AA","BB","CC","DD","EE","FF"]
b = [100,50,75,89,150,200]
plt.bar(a,b,color='red')
```

```python
#BAR GRAPH
a = ["AA","BB","CC","DD","EE","FF"]
b = [100,50,75,89,150,200]
cols123 = ['r','g','b','c','m','y']
plt.bar(a,b,color=cols123)
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.show()
plt.savefig("bar_graph.png") #saves the image as a file
```

```python
a = ["AA","BB","CC","DD","EE","FF"]
b = [100,50,75,89,150,200]
cols123 = ['r','g','b','c','m','k']
plt.barh(a,b,color=cols123)
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.show()
```

```python
# SCATTER PLOT
a = [4,7,8,3,5,6,8,2,1]
b = [5,6,7,8,9,3,1,4,5]
plt.scatter(a,b,marker = "d",s=100,color='red',alpha = 0.8)
plt.axvline(4,color='blue')
plt.axhline(5,color='blue')

# marker styles reference https://matplotlib.org/stable/api/markers_api.html
```
